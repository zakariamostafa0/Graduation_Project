## Split the dataset into classes

```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import sys

labels = pd.read_csv("/split_dataset/SEP-28k_labels.csv")

train_dir ='/split_dataset/dataset/clips/stuttering-clips/clips/'
DR = '/split_dataset/Improved_dataset/Interjection/'
if not os.path.exists(DR):
    os.mkdir(DR)

for filename, class_name in labels.values:
    # Create subdirectory with `class_name`
    if not os.path.exists(DR + str(class_name)):
        os.mkdir(DR + str(class_name))
    src_path = train_dir + '/'+ filename + '.wav'
    dst_path = DR + str(class_name) + '/' + filename + '.wav'
    try:
        shutil.copy(src_path, dst_path)
        print("sucessful")
    except IOError as e:
        print('Unable to copy file {} to {}'
              .format(src_path, dst_path))
    except:
        print('When try copy file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```

## Remove the damage audios

```python

import os
import librosa

# Define the directory containing the audio files
audio_dir = '/split_dataset/Improved_dataset/Interjection/1/'

# Iterate through each file in the directory
for filename in os.listdir(audio_dir):
    # Load the audio file using librosa
    y, sr = librosa.load(os.path.join(audio_dir, filename), sr=None)
    
    # Compute the duration of the audio file
    duration = librosa.get_duration(y=y, sr=sr)
    
    # If the duration is less than one second, delete the file
    if duration < 1:
        print(f'Removing {filename} - duration {duration:.2f}s')
        os.remove(os.path.join(audio_dir, filename))
```

## Remove the redandant audio

```python
import os
import shutil

# Define the source directories
dir1 = "/split_dataset/Improved_dataset/SoundRep/2/"
dir2 = "/split_dataset/Improved_dataset/WordRep/2/"

# Define the destination directory
destination = "/split_dataset/Redantant/2/SoundRep&WordRep/"

# Define the file extension you want to move (e.g., ".txt")
file_extension = ".wav"

# Get a list of all files in the first source directory with the specified file extension
files1 = [f for f in os.listdir(dir1) if f.endswith(file_extension)]

# Loop over the files in the first source directory
for file in files1:
    # Construct the full path to the file in the second source directory
    file2 = os.path.join(dir2, file)
    # Check if the file exists in the second source directory
    if os.path.exists(file2):
        # Construct the full path to the source file in the first directory
        source_file1 = os.path.join(dir1, file)
        # Construct the full path to the source file in the second directory
        source_file2 = os.path.join(dir2, file)
        # Construct the full path to the destination file
        dest_file = os.path.join(destination, file)
        # Move the file to the destination directory
        shutil.copy(source_file1, dest_file)
        os.remove(source_file1)
        os.remove(source_file2)
        
```

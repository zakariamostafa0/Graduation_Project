```
import torch
from transformers import Wav2Vec2Processor, HubertForCTC

processor = Wav2Vec2Processor.from_pretrained("facebook/hubert-large-ls960-ft")
model = HubertForCTC.from_pretrained("facebook/hubert-large-ls960-ft")
    
```

```
import librosa

audio_file_path = "/content/IStutterSoWhat_2_100.wav"

# Load audio file
audio, sample_rate = librosa.load(audio_file_path, sr=16000)
input_values = processor(audio,sampling_rate=sample_rate, return_tensors="pt").input_values

logits = model(input_values).logits
predicted_ids = torch.argmax(logits, dim=-1)
transcription = processor.decode(predicted_ids[0])
```

```
print(transcription)
```

```
! pip install pyspellchecker
```

```
from spellchecker import SpellChecker

# create a SpellChecker object
spell = SpellChecker()



# split the text into words
words = transcription.split()

# create an empty list to store corrected words
corrected_words = []

# iterate over each word in the text
for word in words:
    # if the word is misspelled, replace it with the corrected word
    if not spell.correction(word.lower()) == word.lower():
        corrected_words.append(spell.correction(word.lower()))
    else:
        corrected_words.append(word)

# join the corrected words back into a string
textt = list(filter(None, corrected_words))  # remove None values
corrected_text = " ".join(textt)


print(corrected_text)
```

```
!pip install gTTS

from gtts import gTTS
from IPython.display import Audio


tts = gTTS(corrected_text, lang='en')

tts.save("hello.mp3")
Audio("hello.mp3")
```
